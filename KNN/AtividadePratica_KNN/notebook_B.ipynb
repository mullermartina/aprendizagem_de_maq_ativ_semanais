```python
# Importaçao das bibliotecas
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier # KNeighborsClassifier utiliza dist euclidiana
```

```python
# Carregamento dos dados (dados originais 2 features)
df_orig_2features_testing = pd.read_csv('Dados_Originais_2Features/TestingData_2F_Original.txt', sep='\t')
df_orig_2features_training = pd.read_csv('Dados_Originais_2Features/TrainingData_2F_Original.txt', sep='\t')
```

```python
# Seleção de Features
features_2features = ['total.sulfur.dioxide', 'citric.acid']

label = ['class']
```

```python
# Preparacao dos dados de treino (dados originais 2 features)
x_train_2features_orig = df_orig_2features_training.loc[:, features_2features]
y_train_2features_orig = df_orig_2features_training.loc[:, label].values.ravel()
```

```python
# Preparacao dos dados de teste (dados originais 2 features)
x_test_2features_orig = df_orig_2features_testing.loc[:, features_2features]
y_test_2features_orig = df_orig_2features_testing.loc[:, label].values.ravel()
```

```python
# Modelo KNN com k = 5
knn_classifier_orig_2features = KNeighborsClassifier( n_neighbors= 5 )

# Encaixando o modelo ao Training Dataset
knn_classifier_orig_2features.fit( x_train_2features_orig, y_train_2features_orig )
```

```python
# Selecionando pontos do testing dataset
ponto_teste_N1 = x_test_2features_orig.iloc[[0]]#.values.reshape(1, -1)
ponto_teste_N2 = x_test_2features_orig.iloc[[1]]
ponto_teste_N3 = x_test_2features_orig.iloc[[2]]
ponto_teste_N4 = x_test_2features_orig.iloc[[3]]
```

```python
# Encontrando instâncias mais próximas ao ponto desejado
distancia_N1, indice_N1 = knn_classifier_orig_2features.kneighbors(ponto_teste_N1 , n_neighbors = 5)
distancia_N2, indice_N2 = knn_classifier_orig_2features.kneighbors(ponto_teste_N2 , n_neighbors = 5)
distancia_N3, indice_N3 = knn_classifier_orig_2features.kneighbors(ponto_teste_N3 , n_neighbors = 5)
distancia_N4, indice_N4 = knn_classifier_orig_2features.kneighbors(ponto_teste_N4 , n_neighbors = 5)
```

```python
#Exibição dos resultados: lembrando que "índice" se refere à linha, à instância. "ID" é outra coisa

print(f'Os índices dos 5 vizinhos mais próximos a N1: {indice_N1}') # SEPARAR A QUE T SE REFERE!
print(f'Distâncias dos vizinhos em relação a N1: {distancia_N1}')

print(f'\nOs índices dos 5 vizinhos mais próximos a N2: {indice_N2}') #
print(f'Distâncias dos vizinhos em relação a N2: {distancia_N2}')

print(f'\nOs índices dos 5 vizinhos mais próximos a N3: {indice_N3}') #
print(f'Distâncias dos vizinhos em relação a N3: {distancia_N3}')

print(f'\nOs índices dos 5 vizinhos mais próximos a N4: {indice_N4}') #
print(f'Distâncias dos vizinhos em relação a N4: {distancia_N4}')
```
