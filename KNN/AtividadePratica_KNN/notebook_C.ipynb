```python
# Importaçao das bibliotecas
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier # KNeighborsClassifier utiliza dist euclidiana
from sklearn.metrics import accuracy_score
```

```python
# Encontrando o y previsto para a instância N4

# Carregamento dos dados

# Dados Normalizados 2 Features
df_norm_2features_testing = pd.read_csv('Dados_Normalizados_2Features/TestingData_2F_Norm.txt', sep='\t')
df_norm_2features_training = pd.read_csv('Dados_Normalizados_2Features/TrainingData_2F_Norm.txt', sep='\t')

# Dados Normalizados 11 Features
df_norm_11features_testing = pd.read_csv('Dados_Normalizados_11Features/TestingData_11F_Norm.txt', sep='\t')
df_norm_11features_training = pd.read_csv('Dados_Normalizados_11Features/TrainingData_11F_Norm.txt', sep='\t')

# Seleção de Features
features_2features = ['total.sulfur.dioxide', 'citric.acid']
features_11features = ['fixed.acidity', 'volatile.acidity', 'citric.acid', 'residual.sugar', 'chlorides', 'free.sulfur.dioxide', 'total.sulfur.dioxide', 'density', 'pH', 'sulphates', 'alcohol']

label = ['class']

# Preparacao dos dados de treino
# Dados Normalizados 2 Features
x_train_2features_norm = df_norm_2features_training.loc[:, features_2features]
y_train_2features_norm = df_norm_2features_training.loc[:, label].values.ravel()

# Dados Normalizados 11 Features
x_train_11features_norm = df_norm_11features_training.loc[:, features_11features]
y_train_11features_norm = df_norm_11features_training.loc[:, label].values.ravel()

# Preparacao dos dados de teste
# Dados Normalizados 2 Features
x_test_2features_norm = df_norm_2features_testing.loc[:, features_2features]
y_test_2features_norm = df_norm_2features_testing.loc[:, label].values.ravel()

# Dados Normalizados 11 Features
x_test_11features_norm = df_norm_11features_testing.loc[:, features_11features]
y_test_11features_norm = df_norm_11features_testing.loc[:, label].values.ravel()

# Modelo KNN com k = 5
# Dados Normalizados 2 Features
knn_classifier_norm_2features = KNeighborsClassifier( n_neighbors= 5 )

# Dados Normalizados 11 Features
knn_classifier_norm_11features = KNeighborsClassifier( n_neighbors= 5 )

# Encaixando o modelo ao Training Dataset
# Dados Normalizados 2 Features
knn_classifier_norm_2features.fit( x_train_2features_norm, y_train_2features_norm )

# Dados Normalizados 11 Features
knn_classifier_norm_11features.fit( x_train_11features_norm, y_train_11features_norm )

# Prevendo label para Testing Dataset
# Dados Normalizados 2 Features
y_pred_norm_2features = knn_classifier_norm_2features.predict( x_test_2features_norm )

# Dados Normalizados 11 Features
y_pred_norm_11features = knn_classifier_norm_11features.predict( x_test_11features_norm )

# Comparando y de teste e y previsto
# Dados Normalizados 2 Features
df_norm_2features_testing_N4 = df_norm_2features_testing
df_norm_2features_testing_N4['previsto'] = y_pred_norm_2features
print(f'Resultado obtido para 2 features normalizadas:\n{df_norm_2features_testing_N4}')

# Dados Normalizados 11 Features
df_norm_11features_testing_N4 = df_norm_11features_testing
df_norm_11features_testing_N4['previsto'] = y_pred_norm_11features
print(f'\nResultado obtido para 11 features normalizadas:\n{df_norm_11features_testing_N4[["ID", "citric.acid", "class","previsto"]]}')

```

```python
# Buscando os 5 vizinhos mais próximos a N4
ponto_teste_N4_2features = x_test_2features_norm.iloc[[3]]
ponto_teste_N4_11features = x_test_11features_norm.iloc[[3]]

# Encontrando instâncias mais próximas ao ponto desejado
distancia_N4_2features, indice_N4_2features = knn_classifier_norm_2features.kneighbors(ponto_teste_N4_2features , n_neighbors = 5)
distancia_N4_11features, indice_N4_11features = knn_classifier_norm_11features.kneighbors(ponto_teste_N4_11features , n_neighbors = 5)

print('Para Dados Normalizados 2 Features:')
print(f'Os índices dos 5 vizinhos mais próximos a N4: {indice_N4_2features}')
print(f'Distâncias dos vizinhos em relação a N4: {distancia_N4_2features}')

print('\nPara Dados Normalizados 11 Features:')
print(f'Os índices dos 5 vizinhos mais próximos a N4: {indice_N4_11features}')
print(f'Distâncias dos vizinhos em relação a N4: {distancia_N4_11features}')
```
