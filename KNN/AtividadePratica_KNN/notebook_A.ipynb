```python
# Importaçao das bibliotecas
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier # KNeighborsClassifier utiliza dist euclidiana
from sklearn.metrics import accuracy_score
```

```python
# Carregamento dos dados

# Dados Normalizados 2 Features
df_norm_2features_testing = pd.read_csv('Dados_Normalizados_2Features/TestingData_2F_Norm.txt', sep='\t')
df_norm_2features_training = pd.read_csv('Dados_Normalizados_2Features/TrainingData_2F_Norm.txt', sep='\t')

# Dados Originais 2 Features
df_orig_2features_testing = pd.read_csv('Dados_Originais_2Features/TestingData_2F_Original.txt', sep='\t')
df_orig_2features_training = pd.read_csv('Dados_Originais_2Features/TrainingData_2F_Original.txt', sep='\t')
```

```python
# Seleção de Features
features_2features = ['total.sulfur.dioxide', 'citric.acid']

label = ['class']
```

```python
# Preparacao dos dados de treino
# Dados Normalizados 2 Features
x_train_2features_norm = df_norm_2features_training.loc[:, features_2features]
y_train_2features_norm = df_norm_2features_training.loc[:, label].values.ravel()

# Dados Originais 2 Features
x_train_2features_orig = df_orig_2features_training.loc[:, features_2features]
y_train_2features_orig = df_orig_2features_training.loc[:, label].values.ravel()
```

```python
# Preparacao dos dados de teste
# Dados Normalizados 2 Features
x_test_2features_norm = df_norm_2features_testing.loc[:, features_2features]
y_test_2features_norm = df_norm_2features_testing.loc[:, label].values.ravel()

# Dados Originais 2 Features
x_test_2features_orig = df_orig_2features_testing.loc[:, features_2features]
y_test_2features_orig = df_orig_2features_testing.loc[:, label].values.ravel()
```

```python
# Iniciando modelo KNN com k especificado no exercício
k_valores = [1, 3, 5, 7]

# Dados Normalizados 2 Features
#knn_classifier_norm_2features = KNeighborsClassifier( n_neighbors= k )

# Dados Originais 2 Features
#knn_classifier_orig_2features = KNeighborsClassifier( n_neighbors= k )
```

```python
for k in k_valores:
    # Iniciando modelo KNN com cada k especificado no exercício
    # Dados Normalizados 2 Features
    knn_classifier_norm_2features = KNeighborsClassifier( n_neighbors= k )

    # Dados Originais 2 Features
    knn_classifier_orig_2features = KNeighborsClassifier( n_neighbors= k )

    # Encaixando o modelo ao Training Dataset
    # Dados Normalizados 2 Features
    knn_classifier_norm_2features.fit( x_train_2features_norm, y_train_2features_norm )

    # Dados Originais 2 Features
    knn_classifier_orig_2features.fit( x_train_2features_orig, y_train_2features_orig )

    # Prevendo label para Testing Dataset
    # Dados Normalizados 2 Features
    y_pred_norm_2features = knn_classifier_norm_2features.predict( x_test_2features_norm )

    # Dados Originais 2 Features
    y_pred_orig_2features = knn_classifier_orig_2features.predict( x_test_2features_orig )

    # Avaliando a performance do KNN via acurácia
    # Dados Normalizados 2 Features
    y_test_2features_norm = df_norm_2features_testing.loc[:, label]
    accuracy_norm = accuracy_score( y_test_2features_norm ,  y_pred_norm_2features)
    print(f'\nAcurácia Dados Normalizados com k igual a {k} = {accuracy_norm}')

    # Dados Originais 2 Features
    y_test_2features_orig = df_orig_2features_testing.loc[:, label]
    accuracy_orig = accuracy_score( y_test_2features_orig ,  y_pred_orig_2features)
    print(f'Acurácia Dados Originais com k igual a {k} = {accuracy_orig}')
```
